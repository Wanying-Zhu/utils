### Check if outputs of my code and METAL are the same

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fn_metal_output = '/data100t1/home/wanying/lab_code/utils/meta_analysis/example_output/METAL_meta_output1.tbl'
fn_python_output = '/data100t1/home/wanying/lab_code/utils/meta_analysis/example_output/python_meta_output.meta_output.txt'

df_metal = pd.read_csv(fn_metal_output, sep='\t')
df_python = pd.read_csv(fn_python_output, sep='\t')
df_merged = df_metal.merge(df_python, on='MarkerName')
df_merged
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(10, 4))
ax[0].scatter(df_merged['Zscore_x'], df_merged['Zscore_y'], edgecolors='w')
ax[0].set_title('Zscore')
ax[0].set_xlabel('METAL')
ax[0].set_ylabel("Wanying's code")

ax[1].scatter(df_merged['P-value_x'], df_merged['P-value_y'], edgecolors='w')
ax[1].set_title('P-value')
ax[1].set_xlabel('METAL')
ax[1].set_ylabel("Wanying's code")
```
